```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
background = np.ones((500, 800, 3), dtype=np.uint8)
r = 350

rainbow_list = [(250, 0, 0), #red
                (250, 160, 0), # orange
                 (255, 255, 0), # yellow
                   (0, 240, 0), # green
                      (0, 188, 227), # cyan
                        (0, 0, 245), # indigo
                          (150, 0, 150), # violet
                          (255, 255, 255)] #hite

for i in range(8):
    cv2.circle(background, center = (400, 500), radius = r, color = rainbow_list[i], thickness = -1)
    r -=40

bgr_image = cv2.cvtColor(background, cv2.COLOR_RGB2BGR)
cv2.imwrite('output/rainbow.jpg', bgr_image)

plt.imshow(bgr_image)
plt.show()
```
