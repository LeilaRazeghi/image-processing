```python
import cv2
import numpy as np

# Read the input image
img_watermelon = cv2.imread('input/watermelon.jpg')

# Swap red and green channels where blue is the highest
condition1 = (img_watermelon[:, :, 2] < img_watermelon[:, :, 0]) & (img_watermelon[:, :, 0] > img_watermelon[:, :, 1])
img_matermelon1 = img_watermelon.copy()
img_matermelon1[condition1, 0] = img_watermelon[condition1, 1]
img_matermelon1[condition1, 1] = img_watermelon[condition1, 0]

# Adjust green channel where blue is the highest
condition2 = (img_watermelon[:, :, 2] < img_watermelon[:, :, 1]) & (img_watermelon[:, :, 1] > img_watermelon[:, :, 0])
img_matermelon2 = img_watermelon.copy()
img_matermelon2[condition2, 0] = 255
img_matermelon2[condition2, 1] = np.abs(img_watermelon[condition2, 0] - 20)
img_matermelon2[condition2, 2] = np.abs(img_watermelon[condition2, 2] - 20)

# Save the modified images
#cv2.imwrite('output/matermelon1.jpg', img_matermelon1)
#cv2.imwrite('output/matermelon2.jpg', img_matermelon2)

plt.imshow(img_matermelon1, cmap="gray")
plt.imshow(img_matermelon2, cmap="gray")
```
