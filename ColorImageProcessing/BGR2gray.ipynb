```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
color = cv2.imread("input\photo_2024-03-04_12-37-30.jpg")
color = cv2.cvtColor(color , cv2.COLOR_BGR2RGB)
plt.imshow(color)
```

```python
def rgb2gray(r, g, b):
    gray = ((r + g + b)/3)
    return gray

r, g, b = cv2.split(color)
gray = rgb2gray(r, g, b)

cv2.imwrite('grayphoto.jpg', gray)
plt.imshow(gray, cmap="gray")
```
