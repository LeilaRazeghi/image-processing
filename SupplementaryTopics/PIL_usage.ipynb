```python
from PIL import Image, ImageDraw, ImageFont, ImageOps
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
import numpy as np
```

```python
# read the color image
image = 'input/photo_2024-03-22_00-33-08.jpg'
image = Image.open(image)

```

```python

font_path = 'input\B Nazanin.ttf'
font_size =100
draw = ImageDraw.Draw(image)
font = ImageFont.truetype(font_path, font_size)
text_position = (50, 50)
reshaped_text = arabic_reshaper.reshape('پیکاسو کوچولو')
final_text = get_display(reshaped_text)  
draw.text(text_position, text= final_text, font=font, fill='white')
image.save('output/ashkan_image.png')
```

```python
# calculate and display 3 histogram
r, g, b = image.split()

r_hist = r.histogram()
g_hist = g.histogram()
b_hist = b.histogram()

plt.plot(r_hist, "red")
plt.plot(g_hist, "green")
plt.plot(b_hist, "blue")

plt.title('Original Image Histogram')
plt.show()
```

```python
# Equalize the image histogram
equalized_image = ImageOps.equalize(image)
equalized_image.save('Output/equalized_image.jpg')
plt.imshow(equalized_image)
```

```python
#Calculate and Display the Histogram of the Equalized Image
e_r, e_g, e_b=equalized_image.split()

e_r_hist = e_r.histogram()
e_g_hist = e_g.histogram()
e_b_hist = e_b.histogram()

plt.plot(e_r_hist, "red")
plt.plot(e_g_hist, "green")
plt.plot(e_b_hist, "blue")

plt.title('Equalized Image Histogram')
plt.show()
```

```python
#Convert image to gray.
gray_image = image.convert("L")

gray_image.save("Output/gray_image.jpg")
plt.axis("off")
plt.imshow(gray_image, cmap="gray")
```

```python
#Calculate histogram and show with plt

gray_hist = gray_image.histogram()

plt.plot(gray_hist, color='gray')

plt.title("Gray image histogram")
plt.show()
```

```python
#Equalizes the gray image histogram
equalized_gray_image = ImageOps.equalize(gray_image)

equalized_gray_image.save("Output/equalized_gray_image.jpg")
plt.axis("off")
plt.imshow(equalized_gray_image, cmap='gray')
```

```python
e_gray_hist = equalized_gray_image.histogram()

plt.plot(e_gray_hist, color='gray')

plt.title("Equalized gray image histogram")
plt.show()
```
